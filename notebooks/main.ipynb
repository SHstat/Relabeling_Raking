```python
import numpy as np
import pandas as pd
import sys
sys.path.append("..")

from python.OREM import *
from sklearn.model_selection import train_test_split
```

# Example

```python
df = pd.read_csv("../data/[Conti] compustat.csv", index_col=0)
df.head(3)
```

```python
# 1 should be a label of the minority class
minclass = 1
df_maj = df[df.iloc[:, -1].values != 1]
df_min = df[df.iloc[:, -1].values == 1]

ratio = 0.5
maj_train, maj_test = train_test_split(df_maj, test_size=ratio)
min_train, min_test = train_test_split(df_min, test_size=ratio)

train = pd.concat([maj_train, min_train])
test = pd.concat([maj_test, min_test])
```

```python
label = train.iloc[:, -1]
vc = label.value_counts()
print(vc)

n_maj = vc[0]
n_min = vc[1]
N = n_maj - n_min
```

```python
train = train.iloc[:, :-1].copy()
newDf, newLabel = OREM(train, list(label), N)
newDf
```
