### Destination

```python
import pandas as pd
```

```python
# (# of data)*(# of methods, 7) 
type = "mixed"

wo_AUC_cat = pd.read_csv("../data/RES_"+type+"_exceptAUC.csv")
pastDf = pd.read_csv("../data/"+type+"_preprocessed.csv", index_col=0)
print(pastDf.shape)
display(pastDf.head(8))
```

```python
df = wo_AUC_cat
```

```python
res = []
for clf in ["SVM", "RF", "NB"]:
    for met in ["Gmean", "F1"]:
        target = clf+"_"+met
        tmpDf = df[[c for c in wo_AUC_cat.columns if (target in c)]].mean()
        tmpDf.reset_index(drop=True, inplace=True)
        tmpDf.index = ['OREM'] * len(tmpDf)
        tmpDf.name = met+"_"+clf
        res.append(tmpDf)
        
    dn = [c[:-(len(target)+1)] for c in wo_AUC_cat.columns if (target in c)]
    tmpAuc = auc[[c for c in auc.columns for d in dn if (d in c) & (clf in c)]].mean()
    tmpAuc.reset_index(drop=True, inplace=True)
    tmpAuc.index = ['OREM'] * len(tmpDf)
    tmpAuc.name = "AUC_"+clf
    res.append(tmpAuc)

res = pd.concat(res, axis=1)
```

```python
merged = pd.concat([res.reset_index(drop=True), pastDf.reset_index(drop=True)])
merged.index = list(res.index) + list(pastDf.index)
merged.sort_index(inplace=True)
```
